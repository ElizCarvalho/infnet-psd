```python
#Carregando as lib necessarias
from __future__ import print_function, division
%matplotlib inline
import thinkdsp
import thinkplot
from IPython.display import display
from IPython.display import Audio
```

```python
#Carregando o som do instrumento: Flauta em C
wave_flauta = thinkdsp.read_wave('flute-c.wav')
#Normalizando
wave_flauta.normalize()
#Pegando o segmento de 1seg
segment_flauta = wave_flauta.segment(start=0, duration=1.0)
segment_flauta.make_audio()
```

```python
#Carregando o som do instrumento: Piano em C
wave_piano = thinkdsp.read_wave('piano-c.wav')
#Normalizando
wave_piano.normalize()
#Pegando o segmento de 1seg
segment_piano = wave_piano.segment(start=0, duration=1.0) 
segment_piano.make_audio()
```

```python
#Plot do audio gerado da flauta
segment_flauta.plot()
```

```python
#Plot do audio gerado do piano
segment_piano.plot()
```

```python
#Plot do 'zoom' do segmento de audio gerado da flauta
segment_flauta.segment(start=0, duration=0.005).plot()
```

```python
#Plot do 'zoom' do segmento de audio gerado do piano
segment_piano.segment(start=0, duration=0.005).plot()
```

```python
#Analise espectral do segmento da flauta para descobrir a frequencia fundamental e amplitude
spectrum_flauta = segment_flauta.make_spectrum()
spectrum_flauta.plot(high=400)
```

```python
#Analise espectral do segmento do piano para descobrir a frequencia fundamental e amplitude
spectrum_piano = segment_piano.make_spectrum()
spectrum_piano.plot(high=600)
```

```python
#Analisando em ordem decrescente de amlitude e frequencia. Observa-se que a frequencia e 263hz(corresonde a nota C4)
spectrum_flauta.peaks()[:30]
```

```python
#Harmonicos da frequencia fundamental da flauta (263hz)
for i in range(1,9):
    print(263*i)
```

```python
#Analisando em ordem decrescente de amlitude e frequencia. Observa-se que a frequencia e 527hz(corresonde a nota C5)
spectrum_piano.peaks()[:30]
```

```python
#Harmonicos da frequencia fundamental do piano (527hz)
for i in range(1,9):
    print(527*i)
```

```python
#Recriando o sinal seno da flauta (C4) e plotando
signal_flauta = thinkdsp.SinSignal(freq=263, amp=6960.9)
signal_flauta.plot()
```

```python
#Recriando o sinal seno do piano (C5) e plotando
signal_piano = thinkdsp.SinSignal(freq=527, amp=6383.2)
signal_piano.plot()
```

```python
#Reproduzindo o sinal seno da flauta (C4)
sig_flauta_wave = signal_flauta.make_wave(duration=1.0, start=0, framerate=11025)
audio_flauta = Audio(data=sig_flauta_wave.ys, rate=sig_flauta_wave.framerate)
audio_flauta
```

```python
#Reproduzindo o sinal seno da flauta (C4)
sig_piano_wave = signal_piano.make_wave(duration=1.0, start=0, framerate=11025)
audio_piano = Audio(data=sig_piano_wave.ys, rate=sig_piano_wave.framerate)
audio_piano
```

```python
#Soma dos dois sinais
soma_dos_sinais = signal_flauta + signal_piano
soma_dos_sinais.plot()
```

```python
soma_dos_sinais = signal_piano + signal_flauta
```

```python
#Reproduzindo a soma dos audios - Questao 3
wave_soma_dos_sinais = soma_dos_sinais.make_wave(duration=1.0, start=0, framerate=11025)
wave_soma_dos_sinais.normalize()
wave_soma_dos_sinais.make_audio()
#wave_soma_dos_sinais.write('wave_soma_dos_sinais.wav')
```

```python
#Reproduzindo a soma dos audios - Questao 3
thinkdsp.read_wave('wave_soma_dos_sinais.wav').make_audio()
```

```python
#Recriando o sinal cosseno da flauta (C4) e plotando
cosSignal_flauta = thinkdsp.CosSignal(freq=263, amp=6960.9)

#Recriando o sinal cosseno do piano (C5) e plotando
cosSignal_piano = thinkdsp.CosSignal(freq=527, amp=6383.2)

#Soma dos sinais cossenos
soma_dos_cosSinais = cosSignal_piano + cosSignal_flauta

#Reproduzindo a soma dos cossenos dos áudios 
wave_soma_dos_cosSinais = soma_dos_cosSinais.make_wave(duration=1.0, start=0, framerate=11025)
wave_soma_dos_cosSinais.normalize()
wave_soma_dos_cosSinais.make_audio()
```

```python
#Analise espectral da soma dos sinais de flauta e piano
spectrum_soma_dos_sinais = wave_soma_dos_sinais.segment(start=0, duration=1.0).make_spectrum()
spectrum_soma_dos_sinais.plot(high=600)
```

```python
#Analisando em ordem decrescente de amlitude e frequencia da soma dos senos. 
#Observa-se que a frequencia e 263hz(corresonde a nota C4)
spectrum_soma_dos_sinais.peaks()[:30]
```
