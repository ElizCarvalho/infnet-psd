```python
#Carregando as lib necessarias
from __future__ import print_function, division
%matplotlib inline
import thinkdsp
import thinkplot
import thinkstats2
import numpy as np
from IPython.display import display
from IPython.display import Audio
```

```python
######################## FLAUTA ########################
```

```python
#Origem do Som:
#https://freesound.org/people/fschaeffer/sounds/337884/download/337884__fschaeffer__23flute.wav

#Carregando o som do instrumento: Flauta em C -> Questao 1
wave_flauta = thinkdsp.read_wave('flute-c.wav')
#Normalizando
wave_flauta.normalize()
#Pegando o segmento de 1/2seg
segment_flauta = wave_flauta.segment(duration=0.5)
segment_flauta.make_audio()
```

```python
#Descobrindo o framerate do som coletado da FLAUTA
print('Framerate Flauta: ' + str(segment_flauta.framerate))
```

```python
#Plot do audio gerado da flauta
segment_flauta.plot()
```

```python
#Plot do 'zoom' do segmento de audio gerado da flauta -> Questao 2
segment_flauta.segment(start=0.1,duration=0.025).plot()
```

```python
#Analise espectral do segmento da flauta para descobrir a frequencia fundamental e amplitude -> Questao 3
spectrum_flauta = segment_flauta.make_spectrum()
spectrum_flauta.plot(high=400)
```

```python
#Analisando em ordem decrescente de amlitude e frequencia. 
#Observa-se que a frequencia e 263hz(corresonde a nota C4) -> Questao 3
spectrum_flauta.peaks()[:30]
```

```python
#Harmonicos da frequencia fundamental da flauta (263hz) -> Questao 3
for i in range(1,9):
    print(263*i)
```

```python
#Recriando o sinal seno da flauta (C4) e plotando -> Questao 4
signal_flauta = thinkdsp.SinSignal(freq=263, offset=0)
signal_flauta.plot()
```

```python
#Reproduzindo o sinal seno da flauta (C4) -> Questao 4
sig_flauta_wave = signal_flauta.make_wave(duration=0.5, framerate=96000)
audio_flauta = Audio(data=sig_flauta_wave.ys, rate=sig_flauta_wave.framerate)
audio_flauta
```

```python
#Harmonicas da nota fundamental da flauta  (263hz) -> Questao 3
harmonicos_flauta = []
for i in range(1,9):
    harmonicos_flauta.append(thinkdsp.SinSignal(freq=263*i).make_wave(duration=0.5, framerate=96000))
harmonicos_flauta
```
